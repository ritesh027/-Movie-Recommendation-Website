```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L;

def convert5(obj):
    L = []
    cnt =0
    for i in ast.literal_eval(obj):
        if cnt!=5:         
            L.append(i['name'])
            cnt+=1;
        else:
            break
    return L;

def convertdir(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L;
```

```python
def convertCrew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
            
    return L;
```

```python
movies['crew'] = movies['crew'].apply(convertCrew)
```

```python
# many different words are there with same meaning we need to fix it by stemming it 
# And nltk is a library that helps in stemming the list of words
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer # this is a class
```

```python
maxf=5000
cv = CountVectorizer(max_features=maxf, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
# after getting the feature matrix/(the feature vector of each movie) we need to calculate similarity between the movies
# now we have a 5000 dimensional space and 4806 movies/points and in this space similar movies will have smaller distance 
# our job is to find the simlarity and then tell 5 nearest datapoints to the chosen point / tell 5 similar movies
# here calculating euclidean dist wont be good measure as dimension of data is high i.e. 5000 so we will calculate cosine distance
# cosine distance is the angle b/w to vectors.lesser the angle similar the movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# make a fn
# input =  movie title -> fetch movie index - > in similarity matrix go to that index -> fetch 5 most similar( 5 max values) 
# return the colounm numbers -> fetch the movie data using the coloumn numbers
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1] )[1:7]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie ].index[0];
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1] )[1:7]
    
    for i in movies_list:
       # print(i)
         print(new_df.iloc[i[0]].title)
    
   
```

```python
recommend('The Dark Knight')
```
